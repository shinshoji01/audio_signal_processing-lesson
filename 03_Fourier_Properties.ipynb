```python
import numpy as np
from scipy.signal import triang
from scipy.fftpack import fft
import matplotlib.pyplot as plt
```

```python
x = triang(15)
X = fft(x)
mX = abs(X)
pX = np.angle(X)
```

```python
plt.plot(x)
```

```python
plt.plot(mX)
```

```python
plt.plot(pX)
```

## centered

```python
x = triang(15)
fftbuffer = np.zeros(15)
fftbuffer[:8] = x[7:]
fftbuffer[8:] = x[:7]
X = fft(fftbuffer)
mX = abs(X)
pX = np.angle(X)
```

```python
plt.plot(x)
```
